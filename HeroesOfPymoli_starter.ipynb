## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_items_df = purchase_data.loc[:,["Item ID", "Item Name", "Price"]]
popular_items_df

total_purchase_value = popular_items_df.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Value")
total_purchase_value

average_price = popular_items_df.groupby(["Item ID", "Item Name"]).mean()["Price"].rename("Item Price")
average_price

purchase_count = popular_items_df.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Purchase Count")

summary_data2_df = pd.DataFrame({"Total Purchase Value": total_purchase_value, "Item Price": average_price, "Purchase Count" : purchase_count})
summary_data2_df

summary_data2_df["Total Purchase Value"] = summary_data2_df["Total Purchase Value"].map("${:,.2f}".format)
summary_data2_df["Item Price"] = summary_data2_df["Item Price"].map("${:,.2f}".format)

summary_data2_df = summary_data2_df.sort_values("Purchase Count", ascending=False)
summary_data2_df
```
